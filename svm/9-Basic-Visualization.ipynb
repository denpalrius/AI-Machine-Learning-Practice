## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Line Plot

<p><strong>Explanation:</strong> A line plot is used to display data points connected by a line. Itâ€™s useful for showing trends over time or continuous data.</p>
<p><strong>Usage:</strong> Typically used when you have continuous data, such as changes in a variable over time (e.g., stock prices, temperatures).</p>

```python
# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(8,5))
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot: Sine Wave')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Bar Plot

<p><strong>Explanation:</strong> A bar plot is used to display data in rectangular bars where the height of each bar is proportional to the value it represents.</p>
<p><strong>Usage:</strong> Ideal for comparing quantities across different categories, like sales figures for different products, or populations in different regions.</p>

```python
# Data for Bar plot
categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]

# Create a bar plot
plt.figure(figsize=(8,5))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

## 3. Histogram

<p><strong>Explanation:</strong> A histogram is a graphical representation of the distribution of numerical data. It groups data into bins (intervals) and shows how many data points fall into each bin.</p>
<p><strong>Usage:</strong> Useful for showing the distribution of data (e.g., exam scores, heights of people) and detecting patterns like skewness or outliers.</p>

```python
# Data for histogram
data = np.random.randn(1000)

# Create a histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=1000, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 4. Scatter Plot

<p><strong>Explanation:</strong> A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axes indicates values for an individual data point.</p>
<p><strong>Usage:</strong> Commonly used to determine the relationship or correlation between two variables (e.g., height vs weight).</p>

```python
# Data for scatter plot
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.figure(figsize=(8,5))
plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

## 5. Heatmap (Using Seaborn)

<p><strong>Explanation:</strong> A heatmap is a two-dimensional visualization where individual values in a matrix are represented as colors.</p>
<p><strong>Usage:</strong> Useful for visualizing data density, correlations, or patterns in a dataset (e.g., showing correlations between variables in a correlation matrix).</p>

```python
# Data for heatmap
data = np.random.rand(10, 12)

# Create a heatmap
plt.figure(figsize=(10,7))
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

## 6. Pair Plot (Using Seaborn with a sample dataset)

<p><strong>Explanation:</strong> A pair plot allows you to plot pairwise relationships between different numerical variables in a dataset. It creates a matrix of scatter plots for each combination of variables.</p>
<p><strong>Usage:</strong> Helpful for exploring relationships between multiple variables at once, especially in large datasets (e.g., analyzing correlations in the Iris dataset).</p>

```python
# Load a dataset
iris = sns.load_dataset('iris')

# Create a pair plot
sns.pairplot(iris, hue='species', palette='Set2')
plt.show()
```

### The diagonal elements of the grid show the distribution (density plot) of each feature for the three species.
Example: The top-left plot shows the distribution of sepal length for Setosa, Versicolor, and Virginica.
### The off-diagonal plots represent the scatter plots for each pair of features.
Example: The plot in the second row and first column shows how sepal width relates to sepal length.
### Each point in the scatter plots represents an individual flower sample, with the color indicating the species.

## 7. Pie Chart

<p><strong>Explanation:</strong> A pie chart represents data in a circular graph, where each slice corresponds to a category's contribution to the total.</p>
<p><strong>Usage:</strong> Useful for showing proportional data, like market share, or parts of a whole in percentage.</p>
<pre>

```python
# Pie Chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
plt.show()
```

## 8. Violin Plot

<p><strong>Explanation:</strong> A violin plot combines the features of a box plot and a density plot, showing both the distribution of the data and the probability density. The thicker parts of the violin represent areas where the data points are more concentrated.</p>
<p><strong>Usage:</strong> Useful for comparing the distribution of data across different categories, especially when dealing with multimodal distributions (data with more than one peak).</p>

```python
# Load a dataset
iris = sns.load_dataset('iris')

# violin plot
plt.figure(figsize=(8,5))
sns.violinplot(x="species", y="sepal_length", data=iris, palette="muted")
plt.title("Violin Plot")
plt.show()
```

### 1. Violin Shape:
### The outer shape of each violin represents the density estimate of the data. This shows where the data points are more concentrated or sparse.
Wider parts of the violin indicate areas where data points are more densely packed (higher frequency), while narrower parts indicate lower density (fewer data points).
Box Plot Elements Inside the Violin:

### 2. Black bar (box): This indicates the interquartile range (IQR), which is the range between the 25th and 75th percentiles.
Black lines (whiskers): These extend from the box to show the range of the data, except for outliers.
### 3. Species Comparison:
Each violin represents the distribution of sepal length for one species of iris (Setosa, Versicolor, and Virginica).

### 4. Density and Symmetry:
The shape of each violin helps show whether the data is symmetric or skewed.
For example, Setosa shows a more symmetric distribution, while Versicolor has a slight skew, with more data concentrated towards lower values.

## 9. Swarm Plot

<p><strong>Explanation:</strong> A swarm plot is a categorical scatterplot where the points are adjusted to avoid overlapping, allowing for the visualization of individual data points along with their distribution.</p>
<p><strong>Usage:</strong> Ideal for showing individual data points and how they are distributed across categories. Itâ€™s especially useful when you want to visualize clusters or patterns within the data.</p>

```python
# Load a dataset
iris = sns.load_dataset('iris')

# swarm plot
plt.figure(figsize=(8,5))
sns.swarmplot(x="species", y="sepal_width", data=iris)
plt.title("Swarm Plot")
plt.show()
```

##  A swarm plot is a type of categorical scatter plot where individual data points are represented, and they are adjusted to avoid overlap.

## 10. Area Plot

```python
# Data for area plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,5))
plt.fill_between(x, y1, color="skyblue", alpha=0.5, label="sin(x)")
plt.fill_between(x, y2, color="lightgreen", alpha=0.5, label="cos(x)")
plt.title('Area Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

<p><strong>Explanation:</strong> An area plot is similar to a line plot but with the area between the line and the axis filled in. It helps to visualize the cumulative effect of data over time.</p>
<p><strong>Usage:</strong> Best used when tracking multiple quantities over time and you want to visualize the cumulative effect, such as sales growth, revenue, or any time-series data.</p>

## 11. Hexbin Plot

<p><strong>Explanation:</strong> A hexbin plot displays the relationship between two numerical variables when the data is too dense to plot with individual points. It divides the plot into hexagonal bins and counts the number of points within each bin.</p>
<p><strong>Usage:</strong> Best used for visualizing dense data where individual points would overlap, such as in large datasets with many data points. It helps identify areas of high concentration.</p>

```python
# Data for hexbin plot
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.figure(figsize=(8,5))
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar(label='count in bin')
plt.title('Hexbin Plot')
plt.show()
```

 <p><strong>Explanation:</strong> A hexbin plot displays the relationship between two numerical variables when the data is too dense to plot with individual points. It divides the plot into hexagonal bins and counts the number of points within each bin.</p>
<p><strong>Usage:</strong> Best used for visualizing dense data where individual points would overlap, such as in large datasets with many data points. It helps identify areas of high concentration.</p>


## 12. Pair Grid (Using Seaborn)

<p><strong>Explanation:</strong> A pair grid is an advanced version of a pair plot. It allows you to create a matrix of plots, showing the relationships between variables in a dataset. You can customize the upper, lower, and diagonal parts of the grid to display different types of plots.</p>
<p><strong>Usage:</strong> Useful when you need more control over how different relationships between variables are visualized. Itâ€™s often used in exploratory data analysis to visualize interactions between features in a dataset.</p>

```python
# Pair Grid
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend()
plt.show()
```

## 13. Facet Grid (Using Seaborn)

<p><strong>Explanation:</strong> A facet grid allows you to create multiple subplots, each visualizing a subset of the data based on a particular categorical variable. This allows you to compare trends across different subsets of the data.</p>
<p><strong>Usage:</strong> Best used for visualizing relationships in data subsets. Itâ€™s especially helpful when you have a categorical variable that you want to use to split the data into multiple panels for comparison (e.g., comparing data across different groups or classes).</p>

```python
# Facet Grid
g = sns.FacetGrid(iris, col="species")
g.map(sns.scatterplot, "sepal_length", "sepal_width")
plt.show()
```
