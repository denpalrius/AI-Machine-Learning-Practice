# Image Histogram Analysis using OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

Preview the flower image

```python
img = plt.imread('flower.png')
```

```python
print(img.shape)
```

- **799**: The height of the image (number of pixels along the vertical axis).
- **716**: The width of the image (number of pixels along the horizontal axis).
- **4**: The number of color channels.(Red, Green, Blue, and Alpha), where Alpha represents transparency.

```python
img2 = cv2.imread('flower.png',0)

# find frequency of pixels in range 0-255
histr = cv2.calcHist([img2],[0],None,[256],[0,256])

# show the plotting graph of an image
plt.plot(histr)
```

```python
img3 = cv2.imread('flower.png',0)

# alternative way to find histogram of an image
plt.hist(img3.ravel(),256,[0,256])
plt.show()
```

Load the image in grayscale mode

```python
img_g = cv2.imread('flower.png', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram
histogram = cv2.calcHist([img_g], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.plot(histogram, color='k')  # Black color for grayscale
plt.title('Grayscale Flower Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])
plt.grid()
plt.show()
```

Load the color image

```python
img_c = cv2.imread('flower.png')

# Convert the image from BGR (OpenCV default) to RGB
img_rgb = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)

# Split the image into its Red, Green, and Blue channels
colors = ('r', 'g', 'b')
channel_names = ['Red', 'Green', 'Blue']

# Plot histograms for each channel
plt.figure(figsize=(10, 5))
for i, color in enumerate(colors):
    histogram = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
    plt.plot(histogram, color=color)
    plt.xlim([0, 256])
    plt.title(f'{channel_names[i]} Channel Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Load the image using OpenCV
img_c = cv2.imread('flower.png')

# Convert the image from BGR (OpenCV default) to RGB
img_rgb = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)

# Split the image into its Red, Green, and Blue channels
colors = ('r', 'g', 'b')
channel_names = ['Red', 'Green', 'Blue']

# Create a new figure for the histograms
plt.figure(figsize=(15, 5))

# Number of bins
bins = 256

# Plot histograms for each channel as colored bars in separate subplots
for i, color in enumerate(colors):
    plt.subplot(1, 3, i + 1)  # 1 row, 3 columns, i+1 for the current subplot
    histogram = cv2.calcHist([img_rgb], [i], None, [bins], [0, 256])
    plt.bar(range(bins), histogram[:, 0], color=color, alpha=0.6, width=1)
    
    # Adding titles and labels
    plt.title(f'{channel_names[i]} Channel Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.xlim([0, 256])
    plt.grid()

# Show the plots
plt.tight_layout()
plt.show()
```
